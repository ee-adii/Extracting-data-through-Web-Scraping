# Extracting data through web scraping
#### we are going to extract relevant information from a website 
#### we will BeautifulSoup Lib. to get data

#### Let's GO-

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://keithgalli.github.io/web-scraping/webpage.html'
```

```python
r  = requests.get(url)
```

```python
html_content = r.content
```

```python
soup = bs(html_content)
print(soup)
print(soup.prettify())
```

```python
## geting all the links from this webpage
## links are childrens of socials which is a child of ul (see html body)

anchr = soup.select("ul.socials  a")   ## select('parent child1 child2 ...')

actual_links = [link['href'] for link in anchr]
actual_links
```

```python
## extracting table from webpage

print(soup.prettify())
```

```python
## best way to grab a table is to use pandas lib
## grab all columns and rows in table then 
## buid the dataframe

import pandas as pd
```

```python
tbl = soup.table
tbl
```

```python
## geting columns 
cols_th = tbl.find_all("th")
cols_th
col_name  = [c.string for c in cols_th]
col_name
    
```

```python
## for geting row data(td) 
## first extract all the tr 

table_rows  = soup.table.find('tbody').find_all('tr')
table_rows     ## now we have to extract the td from tr which is our able data
```

```python
## now , from every tr in table_row list extract the td
## then extract all the text from td and strip it because it contains some new line


l = []
for tr in table_rows:
    td  = tr.find_all('td')
    row  = [str(tr.text).strip() for tr in td]
    l.append(row)
    
print(l)     ## this contains all the row data
```

```python
## now the final step is to make dataframe from rows and cols

df = pd.DataFrame(l , columns = col_name)
df  ## we are done hurrey!!
```
